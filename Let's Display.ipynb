```python
@interact
def show_this(Productivity=list(['Scenario 1', 'Scenario 2', 'Scenario 3']), Profitability=list(['Scenario 1', 'Scenario 2', 'Scenario 3']), year=dates):
    temp = pd.DataFrame()
    
    temp_df_display_prod = df_display_prod[df_display_prod['Date (YYYYMM)']==int(year)]
    temp_df_display_prof = df_display_prof[df_display_prof['Date (YYYYMM)']==int(year)]
    
    temp['Legend'] = temp_df_display_prod[Productivity+' Alpha'] + temp_df_display_prof[Profitability+' Alpha']

    
    legend = ['A1', 'A2', 'A3', 'A4', 'B1', 'B2', 'B3', 'B4', 'C1', 'C2', 'C3', 'C4', 'D1', 'D2', 'D3', 'D4', np.nan]
    colors = d3["Category20b"][len(legend)]
    colormap = dict(zip(legend, colors))
    
    temp['Colors'] = [colormap[x] for x in temp.Legend]
    
    source = ColumnDataSource(data = {'Sales_Office' : temp_df_display_prod['Sales Office'], 
                                  'Date_(YYYYMM)': temp_df_display_prod['Date (YYYYMM)'], 
                                  Productivity+'_Prod': temp_df_display_prod[Productivity],
                                  Productivity+'_Alpha_Prod': temp_df_display_prod[Productivity+' Alpha'],
                                  Profitability+'_Prof': temp_df_display_prof[Profitability],
                                  Profitability+'_Alpha_Prof': temp_df_display_prof[Profitability+' Alpha'],
                                  'Legend': temp_df_display_prod[Productivity+' Alpha'] + temp_df_display_prof[Profitability+' Alpha'].map(str),
                                  'Colors': temp['Colors']
                                })
    
    p = figure(plot_height=600, plot_width=750)
    p.xaxis.axis_label = 'Productivity'
    p.yaxis.axis_label = 'Profitability'
    p.xgrid.grid_line_color = None
    p.ygrid.grid_line_color = None
    p.background_fill_color = "#252525"
    
    sp = p.scatter(x= Productivity+'_Prod', y=Profitability+'_Prof', source=source, color = 'Colors', size=6, fill_alpha=0.5)
    
    p.line([10, 10], [9, 41], line_width = 0.50, line_color="lightblue")
    p.line([17.6, 17.6], [9, 41], line_width = 0.50, line_color="lightblue")
    p.line([25.1, 25.1], [9, 41], line_width = 0.50, line_color="lightblue")
    p.line([32.6, 32.6], [9, 41], line_width = 0.50, line_color="lightblue")
    p.line([40, 40], [9, 41], line_width = 0.50, line_color="lightblue")
    p.line([9, 41], [10, 10], line_width = 0.50, line_color="lightblue")
    p.line([9, 41], [17.6, 17.6], line_width = 0.50, line_color="lightblue")
    p.line([9, 41], [25.1, 25.1], line_width = 0.50, line_color="lightblue")
    p.line([9, 41], [32.6, 32.6], line_width = 0.50, line_color="lightblue")
    p.line([9, 41], [40, 40], line_width = 0.50, line_color="lightblue")
    
    
    p.add_tools(HoverTool(renderers=[sp],
        tooltips=[('Sales Office',  '@Sales_Office'),
                  ('Label',  '@Legend')]

        # display a tooltip whenever the cursor is vertically in line with a glyph
        #mode='point'
    ))
    output_notebook()
    show(p)
```

```python
Productivity = widgets.Dropdown(
    options=['Scenario 1', 'Scenario 2', 'Scenario 3'],
    value='Scenario 1',
    description='Productivity',
)

Profitability = widgets.Dropdown(
    options=['Scenario 1', 'Scenario 2', 'Scenario 3'],
    value='Scenario 1',
    description='Profitability',
)

Year = widgets.Dropdown(
    options=dates,
    value='201701',
    description='Year',
)

Submit = widgets.Button(
    description='Submit',
    button_style='success',
    tooltip='Submit',
    icon='check'
)

out = widgets.Output()
display(out)

def on_button_clicked(Submit):
     with out:
        clear_output()
        temp = pd.DataFrame()

        temp_df_display_prod = df_display_prod[df_display_prod['Date (YYYYMM)']==int(Year.value)]
        temp_df_display_prof = df_display_prof[df_display_prof['Date (YYYYMM)']==int(Year.value)]

        temp['Legend'] = temp_df_display_prod[Productivity.value +' Alpha'] + temp_df_display_prof[Profitability.value +' Alpha']


        legend = ['A1', 'A2', 'A3', 'A4', 'B1', 'B2', 'B3', 'B4', 'C1', 'C2', 'C3', 'C4', 'D1', 'D2', 'D3', 'D4', np.nan]
        colors = d3["Category20b"][len(legend)]
        colormap = dict(zip(legend, colors))

        temp['Colors'] = [colormap[x] for x in temp.Legend]

        source = ColumnDataSource(data = {'Sales_Office' : temp_df_display_prod['Sales Office'], 
                                      'Date_(YYYYMM)': temp_df_display_prod['Date (YYYYMM)'], 
                                      Productivity.value +'_Prod': temp_df_display_prod[Productivity.value],
                                      Productivity.value +'_Alpha_Prod': temp_df_display_prod[Productivity.value +' Alpha'],
                                      Profitability.value +'_Prof': temp_df_display_prof[Profitability.value],
                                      Profitability.value+'_Alpha_Prof': temp_df_display_prof[Profitability.value +' Alpha'],
                                      'Legend': temp_df_display_prod[Productivity.value+' Alpha'] + temp_df_display_prof[Profitability.value +' Alpha'].map(str),
                                      'Colors': temp['Colors']
                                    })

        p = figure(plot_height=600, plot_width=750)
        p.xaxis.axis_label = 'Productivity'
        p.yaxis.axis_label = 'Profitability'
        p.xgrid.grid_line_color = None
        p.ygrid.grid_line_color = None
        p.background_fill_color = "#252525"

        sp = p.scatter(x= Productivity.value +'_Prod', y=Profitability.value +'_Prof', source=source, color = 'Colors', size=6, fill_alpha=0.5)

        p.line([10, 10], [9, 41], line_width = 0.50, line_color="lightblue")
        p.line([17.6, 17.6], [9, 41], line_width = 0.50, line_color="lightblue")
        p.line([25.1, 25.1], [9, 41], line_width = 0.50, line_color="lightblue")
        p.line([32.6, 32.6], [9, 41], line_width = 0.50, line_color="lightblue")
        p.line([40, 40], [9, 41], line_width = 0.50, line_color="lightblue")
        p.line([9, 41], [10, 10], line_width = 0.50, line_color="lightblue")
        p.line([9, 41], [17.6, 17.6], line_width = 0.50, line_color="lightblue")
        p.line([9, 41], [25.1, 25.1], line_width = 0.50, line_color="lightblue")
        p.line([9, 41], [32.6, 32.6], line_width = 0.50, line_color="lightblue")
        p.line([9, 41], [40, 40], line_width = 0.50, line_color="lightblue")


        p.add_tools(HoverTool(renderers=[sp],
            tooltips=[('Sales Office',  '@Sales_Office'),
                      ('Label',  '@Legend')]

            # display a tooltip whenever the cursor is vertically in line with a glyph
            #mode='point'
        ))
        output_notebook()
        show(p)

Submit.on_click(on_button_clicked)

display(Productivity) 
display(Profitability)
display(Year)
display(Submit)
```
